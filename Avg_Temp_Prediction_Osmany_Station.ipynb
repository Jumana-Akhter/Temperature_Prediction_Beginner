```python
from sklearn.linear_model import Ridge
reg = Ridge(alpha=.1)
```

```python
predictors=["precip","tavg","tmax","tmin"]
```

```python
train = weather_Osmany_point.loc[:"2020-12-31"]
```

```python
test = weather_Osmany_point.loc["2021-01-01":]
```

```python
reg.fit(train[predictors],train["target"])
```

```python
combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1) 
combined.columns = ["actual","predictions"]
```

```python
def create_predictions(predictions, weather_Osmany_point, reg):
    train = weather_Osmany_point.loc[:"2020-12-31"]
    test = weather_Osmany_point.loc["2021-01-01":]
    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])
    error = mean_absolute_error(test["target"],predictions)
    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["actual","predictions"]
    return error, combined 
```

```python
weather_Osmany_point["month_max"] = weather_Osmany_point["tavg"].rolling(30).mean()
```

```python
weather_Osmany_point
```

```python
weather_Osmany_point["mon_max"] = weather_Osmany_point["month_max"]/weather_Osmany_point["tavg"]
```

```python
weather_Osmany_point["max_min"] = weather_Osmany_point["tmax"]/weather_Osmany_point["tmin"]
```

```python
predictors=["precip","tavg","tmax","tmin","month_max","max_min"]
```

```python
weather_Osmany_point = weather_Osmany_point.iloc[30:,:].copy()
```

```python
error, combined = create_predictions(predictors, weather_Osmany_point, reg)
```

```python
monthly_avg = weather_Osmany_point["tavg"].groupby(weather_Osmany_point.index.month).apply(lambda x: x.expanding(1).mean())
weather_Osmany_point["mon_avg"] = monthly_avg.reset_index(level=0, drop=True) 
```

```python
day_of_year_average = weather_Osmany_point["tavg"].groupby(weather_Osmany_point.index.day_of_year).apply(lambda x: x.expanding(1).mean())
weather_Osmany_point["day_of_year_avg"] = day_of_year_average.reset_index(level=0, drop=True)
```
